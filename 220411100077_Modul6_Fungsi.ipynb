## 2.1 Cek List
Buatlah Fungsi dan code untuk memanggil fungsi dengan menggunakan Python, antara lain :
- Fungsi createList untuk membentuk suatu list dengan parameter atau argumen berupa ukuran list
- Fungsi isPrime untuk menentukan apakah suatu bilangan adalah bilangan prima, dengan argumen
adalah bilangan yang akan dicek
- Fungsi createListPrime untuk membuat suatu list yang berisi bilangan prima, dengan ketentuan :

     1. argumen atau paramater adalah list dimana di setiap anggotanya akan dicek termasuk bilangan prima ataukah tidak
     2. gunakan fungsi isPrime yang sudah dibuat, untuk cek apakah suatu bilangan adalah bilangan prima
     3. anggota-anggota list yang dihasilkan dari fungsi ini,  tidak boleh sama [Hint: gunakan in]

Contoh output yang dihasilkan dapat dilihat pada Gambar 1.

```python
# jawaban
def createList(temp):
    data=[]
    for i in range(temp):
        inp=int(input(f"masukkan data ke- {i} : "))
        data.append(inp)  
    return data
def isPrime(b):
    temp=0
    for i in range(1,b+1):
        if b%i==0:
            temp=temp+1
    if temp==2:
        return True 
    else:
        return False       
def createListPrime(jamak):
    prima=[]
    for i in jamak:
        if isPrime(i) and i not in prima:
            prima.append(i)
    return prima    
```

```python
data=createList(10)
print('data list=',data)
prima=createListPrime(data)
print('data prima =',prima)
```

# 2.2	Operasi List
Buatlah Fungsi dan code untuk memanggil fungsi dengan menggunakan Python, antara lain :
-	Fungsi createList untuk membentuk suatu list dengan parameter atau argumen berupa ukuran list
-	Fungsi avgList untuk menghitung rata-rata suatu list, dengan argumen adalah list
-	Fungsi addList untuk menjumlahkan dua buah list yang memiliki ukuran yang sama Contoh output 
yang dihasilkan dapat dilihat pada Gambar 2.


```python
# jawaban
def createList(angka):
    list1=[]
    for i in range(angka):
        temp=int(input(f"masukkan data ke-{i} : "))
        list1.append(temp)
    return list1
def avgList(list1):
    total=0
    for i in range(len(list1)):
        total=total+list1[i]
        ratarata=total/len(list1)
    return ratarata 
def addList(a,b):
    list1=[]
    for i in range(len(a)):
        if len(a)==len(b):
            total=a[i]+b[i]
            list1.append(total)
        else:
          list1="ukuran list tidak sama"
    return list1
```

```python
data1=createList(3)
print('data-1 =',data1,' ; rata-rata=',avgList(data1))
data2=createList(3)
print('data-2 =',data2,' ; rata-rata=',avgList(data2))
hasil=addList (data1,data2)
hasil2=addList(hasil,hasil)
print(data1,'+',data2,'=',hasil)
print(hasil,'+',hasil,'=',hasil2)
```

```python
# jawaban
def koordinatY(x,a,b):
    return [x[i]*a+b for i in range(len(x))]
```
