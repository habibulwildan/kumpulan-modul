## 2.1 Program Bilangan ganjil
Konversi algoritma yang sudah ditulis sebelumnya dengan menggunakan bahasa Pemrograman Python. Contoh hasil eksekusi dapat dilihat di Gambar.2a or Gambar.2b :



```python
# Jawaban
n=10
pemberhentian=0
counter=1
temp=0
while counter<=n:
    if pemberhentian%2==1:
        print('bilangan ke-',counter,':',pemberhentian)
        counter+=1
        temp=temp+pemberhentian
    pemberhentian+=1
print('Total','=',temp)
```
